### Cleaning Data

```python
#GEOLOCATION
#Menghapus duplikasi data
geolocation_df.drop_duplicates(inplace=True)
print("Jumlah duplikasi data: ", geolocation_df.duplicated().sum())
```

```python
#ORDER ITEM
#mengubah tipe data menjadi datetime
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])
order_items_df.info()
```

```python
#ORDER REVIEW
#Mengubah tipe data menjadi datetime
datetime_columns = ["review_creation_date", "review_answer_timestamp"]

for column in datetime_columns:
  order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])

order_reviews_df.info()
```

```python
#Mengatasi missing value
order_reviews_df[order_reviews_df.review_comment_title.isna()]
order_reviews_df.review_comment_title.value_counts()

#Mengisi missing value dengan string tetap yakni "No Title", "No Review"
order_reviews_df.review_comment_title.fillna("No Title", inplace=True)
order_reviews_df.review_comment_message.fillna("No Review", inplace=True)

order_reviews_df.isna().sum()
```

```python
#ORDER
#Mengubah tipe data menjadi date
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
#PRODUCT
#Mengatasi missing value
products_df.dropna(inplace=True)
products_df.isna().sum()
```

**Insight:**
* Tidak ada lagi data yang duplikat
* Merubah tipe data shipping_limit_date menjadi date
* Merubah tipe data review_creation_date dan review_answer_timestamp menjadi date
* Mengisi missing value pada tabel order review dengan "No title" untuk kolom review_comment_title dan "No Review" untuk kolom review_comment_message
* Merubah tipe data order_purchase_timestamp, order_approved_at, order_delivered_carrier_date, order_delivered_customer_date, order_estimated_delivery_date menjadi date
* Menghapus missing value pada tabel product

```python
#menggabungkan tabel order_items_payments dan order_orders_reviews_
all_orders_df = pd.merge(
    left=order_items_payments,
    right=order_orders_reviews,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_orders_df.head()
```

```python
#Menghitung lamanya pengiriman pada (status = delivered)
delivered_orders = all_orders_df[all_orders_df['order_status'] == "delivered"]

# Hitung delivery_time
delivered_orders['delivery_time'] = delivered_orders["order_delivered_customer_date"] - delivered_orders["order_purchase_timestamp"]

# Konversi delivery_time menjadi hari (total detik dibagi 86400 detik per hari)
delivered_orders["delivered_time"] = delivered_orders['delivery_time'].dt.total_seconds() / 86400

# Tambahkan kolom delivered_time ke all_orders_df
all_orders_df.loc[delivered_orders.index, "delivered_time"] = delivered_orders["delivered_time"]

# Tampilkan hasil
print(all_orders_df[["order_id", "order_status", "order_delivered_customer_date", "order_purchase_timestamp", "delivered_time"]])
```

```python
#Mengetahui rata-rata delivered time berdasarkan review score
delivery_vs_rating = all_orders_df.groupby('review_score')['delivered_time'].mean()
print(delivery_vs_rating)
```

```python
#Mengecek bagaimana ulasan produk (review score)
all_orders_df[all_orders_df.order_status == "delivered"].groupby("review_score").order_id.count().sort_index(ascending=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana demografi pelanggan yang kita miliki?

```python
#Visualisasi demografi customer berdasarkan city
bycity_df = customers_df.groupby(by="customer_city").customer_id.nunique().reset_index()
bycity_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

plt.figure(figsize=(10, 5))

sns.barplot(
    y="customer_count",
    x="customer_city",
    data=bycity_df.sort_values(by="customer_count", ascending=False).head(5)
)
plt.title("Number of Customer by city", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

```python
#Visualisasi demografi customer berdasarkan state
bystate_df = customers_df.groupby(by="customer_state_full").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

plt.figure(figsize=(10, 5))

sns.barplot(
    y="customer_count",
    x="customer_state_full",
    data=bystate_df.sort_values(by="customer_count", ascending=False).head(5)
)
plt.title("Number of Customer by state", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 2: Bagaimana tren berdasarkan waktu pembelian pelanggan?

```python
trend_by_time.index = trend_by_time.index.to_timestamp()

# Membuat line chart
plt.figure(figsize=(10, 6))
plt.plot(trend_by_time.index, trend_by_time.values, marker='o', color='blue', label='Jumlah Pesanan Delivered')

# Menambahkan label dan judul
plt.title('Tren Pesanan Delivered Berdasarkan Waktu', fontsize=16)
plt.xlabel('Waktu (Bulan-Tahun)', fontsize=12)
plt.ylabel('Jumlah Pesanan Delivered', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)  # Memutar label sumbu X untuk keterbacaan
plt.legend()
plt.tight_layout()

# Menampilkan chart
plt.show()
```

### Pertanyaan 3: Produk apa yang paling banyak dan paling sedikit terjual?

```python
sum_order_items_df = all_df.groupby("product_category_name_english").order_id.count().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="order_id", y="product_category_name_english", data=sum_order_items_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="order_id", y="product_category_name_english", data=sum_order_items_df.sort_values(by="order_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Best and Worst Performing Product by Number of Sales", fontsize=20)
plt.show()
```

### Pertanyaan 4: Bagaimana tren berdasarkan ulasan pelanggan terhadap produk?

```python
# Filter data untuk order yang berstatus 'delivered'
byreview_df = all_orders_df[all_orders_df.order_status == "delivered"] \
    .groupby("review_score").order_id.count().reset_index()

# Ganti nama kolom agar lebih sesuai
byreview_df.rename(columns={
    "order_id": "order_count"
}, inplace=True)

# Plot barplot
plt.figure(figsize=(10, 5))

sns.barplot(
    y="order_count",
    x="review_score",
    data=byreview_df.sort_index(ascending=True)
)
plt.title("Number of Delivered Orders by Review Score", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel("Review Score", fontsize=12)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 5: Metode pembayaran apa yang paling banyak digunakan?

```python
#Visualisasi metode pembayaran yang pling banyak digunakan
filtered = all_orders_df[all_orders_df.order_status == "delivered"]
payment_df = filtered.groupby(by="payment_type").customer_id.count().reset_index()
payment_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

plt.figure(figsize=(10, 5))

sns.barplot(
    y="customer_count",
    x="payment_type",
    data=payment_df.sort_values(by="customer_count", ascending=False)
)
plt.title("Number of Payment Method", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 6: Berapa rata-rata waktu pengiriman barang di tiap states?

```python
state_delivery_time = delivered_time.groupby('customer_state').delivered_time.mean().reset_index()

# Membuat bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='customer_state', y='delivered_time', data=state_delivery_time, palette='viridis')

# Menambahkan label dan judul
plt.title('Rata-rata Waktu Pengiriman Berdasarkan State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Rata-rata Waktu Pengiriman (Hari)', fontsize=12)
plt.xticks(rotation=45)  # Memutar label sumbu X untuk keterbacaan
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Menampilkan chart
plt.show()
```
